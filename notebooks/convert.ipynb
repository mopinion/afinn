```python
import json
```

```python
data = json.loads(open('../afinn/data/AFINN-fi.json').read())
```

```python
loc = '../AFINN/data/AFINN-fi-165.txt'
txt = ''
for word, score in data.items():
    print(word)
    print(score)
    txt += '\n{}\t{}'.format(word, score)
print(txt)
f = open(loc, 'w').write(txt)
```
